```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Total Amount'])
plt.title("Total Sales Distribution")
plt.show()
```

## **Total Sales Distribution**

### **Insight:**
The histogram of 'Total Sales Distribution' shows a right-skewed distribution, indicating that the majority of transactions have lower total amounts, while a smaller number of transactions account for very high total sales. The distribution is characterized by a long tail extending towards higher values. This suggests that while most customers make smaller purchases, there are occasional high-value transactions that significantly contribute to the overall sales. Companies might consider strategies to encourage more frequent high-value purchases or to increase the average transaction value for the majority of customers.

```python
df.groupby('Product Category')['Total Amount'].sum().plot(kind='bar')

plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.show()
```

# **Sales By Product**
## **Insight**:
The chart shows that Electronics has the highest total sales, followed closely by Clothing, while Beauty has the lowest sales among the three categories. This suggests that customers purchase more electronics and clothing products compared to beauty items. The company may focus on promoting beauty products to increase their sales.

```python
sns.barplot(x='Gender', y='Total Amount', data=df)

plt.title("Sales by Gender")
plt.show()
```

# **Sales  By Gender**
# **Insight:**
Sales from male and female customers are almost equal, with females contributing slightly more. This indicates no significant gender-based difference in purchasing behavior.

```python
df['Date'] = pd.to_datetime(df['Date'])

sales_trend = df.groupby('Date')['Total Amount'].sum()

sales_trend.plot(figsize=(10,5))

plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Total Amount', data=df)

plt.title("Age vs Total Sales")
plt.show()
```

# **Age vs Total Sales Scatter Plot**

## **Insight:**
The scatter plot illustrates the relationship between customer age and total purchase amount. The points are widely distributed across different age groups, indicating no strong relationship between age and total sales. Customers from various age ranges appear to spend similar amounts, suggesting that purchasing behavior is relatively consistent across age groups.

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

### **Correlation Heatmap**

# **Insight:**
The correlation heatmap shows the relationships between numerical variables in the dataset. Price per Unit and Total Amount have a strong positive correlation (0.85), indicating that higher product prices significantly increase the total sales amount.
Quantity and Total Amount show a moderate positive correlation (0.37), meaning purchasing more items also increases total sales. Other variables such as Age and Transaction ID show very weak correlations with sales, suggesting they have little direct impact on the total purchase value.




## **Conclusion**
The exploratory data analysis of retail sales data reveals that most transactions involve lower purchase amounts, with a few high-value transactions contributing significantly to overall sales. Among product categories, electronics and clothing generate the highest revenue, while beauty products lag behind. Sales are almost equally distributed between male and female customers, showing no major gender-based difference. Additionally, factors like age do not have a strong influence on purchasing behavior. Price per unit plays a significant role in determining total sales, with quantity also contributing moderately.
